Cosine wave

```python
import numpy as np
t = np.arange(1, 2, 0.01)

#matplotlib inline --> working jupyter notebook bcs graph not showing, totally hide

a = 1+np.cos(2*np.pi*t) # amplitude formula
plt.plot(t, a, '*')  # this graph is the following to the + design
# plt.plot(t, a, '--') # graph design is --
# plt.plot(t, a, '++') # graph design is --
# plt.plot(t, a, 'val') # graph design is val type design

plt.grid() # this is showing horizontal, vertical shape
plt.xlabel("Time(t)") # x axis line
plt.ylabel("Amplitude(A)") # y axis line
plt.title("Cosine wave graph") # to the graph name

plt.show()
```

## Bar Graph

```python
x = [1, 2, 3, 4] # x axis value
y = [23, 45, 56, 78] # y axis value
label = ['Fail', 'Second Division', 'First Division', 'Distinction'] # x axis defines

plt.bar(x, y, tick_label=label, width=0.5, color='Green')
plt.xlabel('x-axis') # x axis
plt.ylabel('y-axis') # y axis
plt.title('Bar graph') # showing about the graph name
plt.show()
```

## Pie Chart

```python
contents = ['Animals', 'Plants', 'Human'] # different type name
slices = [3, 7, 9] # every slices how to range
colors = ['green', 'orange', 'pink'] # easily findout to the graph

plt.pie(slices, labels=contents, colors=colors) # defines three variable
plt.show()
```

```python
contents = ['Animals', 'Plants', 'Human'] # different type name
slices = [3, 7, 9] # every slices how to range
colors = ['green', 'orange', 'pink'] # easily findout to the graph

plt.pie(slices, labels=contents, colors=colors, startangle=90) # defines three variable, star tangle is 90 degree
plt.show()
```

```python
contents = ['Animals', 'Plants', 'Human'] # different type name
slices = [3, 7, 9] # every slices how to range
colors = ['green', 'orange', 'pink'] # easily findout to the graph

plt.pie(slices, labels=contents, colors=colors, startangle=90, shadow = True, autopct ='%1.3f%%')
# defines three variable, star tangle is 90 degree, shadow and auto percentage is showing the graph
plt.show()
```

```python
contents = ['Animals', 'Plants', 'Human'] # different type name
slices = [3, 7, 9] # every slices how to range
colors = ['green', 'orange', 'pink'] # easily findout to the graph

plt.pie(slices, labels=contents, colors=colors, startangle=90, shadow = True, autopct ='%1.3f%%', explode=(0.05, 0, 0))
# defines three variable, star tangle is 90 degree, shadow and auto percentage is showing the graph
plt.legend() # the graph mentioned about different color showing
plt.show()
```

## Histogram(it looks like bar graph)

```python
percentage = [25, 45, 15, 25, 78, 89, 54, 12, 23, 65, 81, 32, 27, 39, 37, 29, 94]
range = (0, 100)
num = 5

# Change the range argument to a tuple with two elements: (0, 100) in range
plt.hist(percentage, num, range, color='orange', histtype='bar', rwidth=0.5)

plt.xlabel('percentage')
plt.ylabel('students')
plt.title('Histogram plot')  # Corrected the function name

plt.show()
```
